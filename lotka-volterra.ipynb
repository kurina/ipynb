\begin{align}
\frac{dx}{dt} & = \alpha x - \beta x y \\[6pt]
\frac{dy}{dt} & = \delta x y  - \gamma y
\end{align}

```python
from numpy import *
import pylab as p
from ipywidgets import *
from scipy.integrate import *
%matplotlib inline


alfa= 1.0   # Growing rate of prey when there's no predators
beta= 0.1   # Dying rate of prey due to predation
gama=1.0    # Dying rate of predators when there's no prey
delta=0.075  # Reproduction rate of predators per 1 prey eaten


#X0=array([Ovce,Vlci])  # pocet pred a predatorov
period=1000
time=linspace(0,15,period)
def lotka_volterra(initial,t):
    x=initial[0]
    y=initial[1]
    dx_dt=alfa*x-beta*x*y
    dy_dt=delta*x*y-gama*y
    return array([dx_dt,dy_dt])

def sim(Ovce,Vlci,beta,delta):    
    X0=array([Ovce,Vlci])
    X,infodict=odeint(lotka_volterra, X0, time, full_output=True)
    #o,v=X[:,0],X[:,1]
    #alebo
    o,v=X.T
    p.plot(time,o, 'b-', label='Ovce')
    p.plot(time,v,'r-', label='Vlci')
    p.legend()
    
interactive(sim,Ovce=(1,20,1),Vlci=(1,10,1),beta=(0.1,1.,.1),delta=(0.03,0.1,0.01))
```
