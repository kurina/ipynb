```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline

def g(y,x):
    #print "Y= ",y,"X= ",x
    y0 = y[0]
    y1 = y[1]
    y2 = (6*x-8)*y0/(3*x-1)+(3*x+2)*y1/(3*x-1)
    return [y1,y2]

init = [2.0,3.0]
time=np.linspace(-2,2,100)
a=odeint(g,init,time)
#print sol
plt.plot(time,a[:,0])
#plt.plot(time,b)
plt.show()
```


\begin{align}
{dx \over dt} = rx\left(1-{x \over K}\right)
\end{align}


```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from numpy import linspace
%matplotlib inline


def dx_dt(y,x):
    return [y[0]*x*(1-x/y[1]),x]

r=0.75
K=0.01
y0=np.array([r,K])
times = linspace(0.0, 5.0, 100)

sol=odeint(dx_dt, y0, times)
out=sol.T
plt.plot(out[0],times)
plt.plot(out[1],times)
plt.show()
```

```python
import scipy.integrate as integrate
import matplotlib.pyplot as plt
import numpy as np
import pylab
%matplotlib inline

pi = np.pi
sqrt = np.sqrt
cos = np.cos
sin = np.sin

def deriv_z(z, phi):
    u, udot = z
    return [udot, -u + sqrt(u)]

phi = np.linspace(0, 7.0*pi, 2000)
zinit = [1.49907, 0]
z = integrate.odeint(deriv_z, zinit, phi)
u, udot = z.T
# plt.plot(phi, u)
fig, ax = plt.subplots()
ax.plot(1/u*cos(phi), 1/u*sin(phi))
ax.set_aspect('equal')
plt.grid(True)
plt.show()
```
